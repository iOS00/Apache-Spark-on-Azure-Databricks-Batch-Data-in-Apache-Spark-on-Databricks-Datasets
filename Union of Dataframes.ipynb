```
customers_australia_df = spark.read.format("csv") \
                              .option("inferSchema", True) \
                              .option("header", True) \
                              .option("sep", ",") \
                              .load("/FileStore/datasets/customers_australia.csv")
display(customers_australia_df)
```

```
customers_israel_df = spark.read.format("csv") \
                              .option("inferSchema", True) \
                              .option("header", True) \
                              .option("sep", ",") \
                              .load("/FileStore/datasets/customers_israel.csv")
display(customers_israel_df)
```

```
# the schema of each above dataframe is identical (the same number of columns and their types). This is must condition for Union dataframes
union_customer_df = customers_israel_df.union(customers_australia_df)
display(union_customer_df)
```
