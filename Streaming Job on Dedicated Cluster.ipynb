While our job was created and is running streming job(by executing sript in .py file) on dedicated Job cluster, we will query the streaming data output of the job in this notebook (attached to our primary cluster)

```
# load all .csv files from a destination location
attr_stream_data = spark.readStream.format("cloudFiles") \
                            .option("cloudFiles.format", "csv") \
                            .option("header", "False") \
                            .option("cloudFiles.schemaLocation", 
                                    "dbfs:/FileStore/datasets/attr_source_stream") \
                            .load("dbfs:/FileStore/datasets/attr_dest_location/*.csv")
attr_stream_data.display()
```

Now remember to cancel Job run (needs to be done manually for jobs without triggers, activated manually). After job is terminated, the dedicated job cluster will be terminated also. But you will still be able to observe job's logs and history after job's termination
